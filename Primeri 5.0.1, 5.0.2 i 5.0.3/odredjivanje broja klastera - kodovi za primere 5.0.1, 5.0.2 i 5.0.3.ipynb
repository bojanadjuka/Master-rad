```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn import mixture
import math 
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from PIL import Image
```

### Primer 5.0.1.

```python
X, y = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=0)
X = X[:, ::-1] 
```

```python
plt.figure(figsize=(5,3))
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
scaler=StandardScaler()
scaler.fit(X)
```

```python
X_scaled=scaler.transform(X)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1])
plt.show()
```

```python
silhoutte=[]
bic=[]
aic=[]
K=range(2,7)
for k in K:
    model = mixture.GaussianMixture(n_components=k)
    model.fit(X_scaled)
    labels = model.predict(X_scaled)
    silhoutte.append(metrics.silhouette_score(X_scaled, labels))
    bic.append(model.bic(X_scaled))
    aic.append(model.aic(X_scaled))
```

```python
img_array=np.asarray(img).astype('float32')
img_array=img_array.reshape((-1,3))

fig = plt.figure(figsize=(15,15))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(img_array[:,0],img_array[:,1],img_array[:,2],s=0.01)
ax.set_xlabel('$R$', fontsize=10)
ax.set_ylabel('$G$', fontsize=10)
ax.set_zlabel('$B$', fontsize=10)
ax.zaxis.set_tick_params(labelsize=10)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)

plt.show()
```

```python
silhoutte_score2=[]
bic2=[]
aic2=[]
K=range(2,10)
for k in K:
    model = mixture.GaussianMixture(n_components=k)
    model.fit(img_array/255)
    labels = model.predict(img_array/255)
    silhoutte_score2.append(metrics.silhouette_score(img_array/255, labels))
    bic2.append(model.bic(img_array/255))
    aic2.append(model.aic(img_array/255))
    
```
