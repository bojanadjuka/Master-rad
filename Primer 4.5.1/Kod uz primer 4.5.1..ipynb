```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import mixture
from sklearn.preprocessing import StandardScaler
```

```python
x1=list(np.random.choice(400, 200, replace=True))+list(np.random.choice(range(250,400), 150, replace=True))+list(np.random.choice(range(470,1100), 200, replace=True))
```

```python
x2=list(np.random.choice(range(20,50), 200, replace=True))+list(np.random.choice(range(70,150), 150, replace=True))+list(np.random.choice(range(60,110), 200, replace=True))
```

```python
x1=[x+50*random.random()-random.random() for x in x1]
x2=[x+1+30*random.random()-random.random() for x in x2]
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x1, x2)
plt.show()
```

```python
y=200*[1]+150*[3]+200*[2]
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x1, x2,  c=y, s=40, cmap='cool')
plt.show()
```

```python
plt.figure(figsize=(10,3))

plt.subplot(1,2,1)
plt.scatter(x1, x2)
plt.title('  ',fontsize=9)

plt.subplot(1,2,2)
plt.scatter(x1, x2,  c=y, s=40, cmap='cool')
plt.title('    ',fontsize=9)

plt.show()
```

```python
plt.figure(figsize=(10,3))

plt.subplot(1,2,1)
plt.scatter(x1, x2)

plt.subplot(1,2,2)
plt.scatter(x1, x2,  c=y, s=40, cmap='cool')

plt.show()
```

```python
X=np.array(list(zip(x1, x2)))
```

```python
scaler=StandardScaler()
scaler.fit(X)
X=scaler.transform(X)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(X[:,0], X[:,1],  c=y, s=40, cmap='plasma')
plt.show()
```

```python
model_gmm=mixture.GaussianMixture(n_components=3)
```

```python
model_gmm.fit(X)
labels_gmm = model_gmm.predict(X)
```

```python
model_kmeans=KMeans(n_clusters=3, random_state=0, n_init="auto").fit(X)
```

```python
labels_kmeans=model_kmeans.labels_
```

```python
plt.figure(figsize=(5,3))
plt.scatter(X[:,0], X[:,1],  c=labels_kmeans, s=40, cmap='cool')
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.scatter(X[:,0], X[:,1],  c=labels_gmm, cmap='cool')
plt.show()
```

```python
plt.figure(figsize=(10,3))

plt.subplot(1,2,1)
plt.scatter(X[:,0], X[:,1],  c=labels_kmeans, s=40, cmap='cool')
plt.title('    ',fontsize=9)

plt.subplot(1,2,2)
plt.scatter(X[:,0], X[:,1],  c=labels_gmm, cmap='cool')
plt.title('     ',fontsize=9)

plt.show()
```
