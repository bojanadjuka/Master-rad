```python
import numpy as np
import matplotlib.pyplot as plt
import math 
import random
from scipy.stats import norm 
```

```python
mu1, sigma1 = 5, 4
normal1 = np.random.normal(mu1, sigma1, 1000)
```

```python
mu2, sigma2 = 10, 2 
normal2 = np.random.normal(mu2, sigma2, 1000)
```

```python
def generisanje_uzorka(n_uzorka,mu,sigma,pi,n_mesavina):
    uzorak=[]
    which_mixture= random.choices(range(n_mesavina), weights=pi, k=n_uzorka)
    for i in which_mixture:
        uzorak.append(np.random.normal(mu[i], sigma[i], 1)[0])
    return(uzorak)
    
```

```python
x_axis=np.linspace(start=-10,stop=20,num=5000)
```

```python
n_uzorka=5000
mu=[5,10]
sigma=[4,2]
pi=[0.5,0.5]
n_mesavina=2
plt.figure(figsize=(11,4))
plt.hist(generisanje_uzorka(n_uzorka,mu,sigma,pi,n_mesavina),bins=100,density=True)
plt.plot(x_axis, norm.pdf(x_axis, mu[0], sigma[0]),c='red',label='  N(5,16)',linewidth=2)
plt.plot(x_axis, norm.pdf(x_axis, mu[1], sigma[1]),c='lime',label='  N(10,4)',linewidth=2)
plt.plot(x_axis,pi[0]*norm.pdf(x_axis, mu[0], sigma[0])+pi[1]*norm.pdf(x_axis, mu[1], sigma[1]),c='black',label='    ',linewidth=2)

plt.legend(fontsize='x-small')
plt.grid()
plt.show()
```
