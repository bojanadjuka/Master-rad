```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from scipy import stats
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn import mixture
import math 
from random import sample
from sklearn.preprocessing import StandardScaler
```

```python
X, y_true = make_blobs(n_samples=400, centers=2, cluster_std=0.8, random_state=0)
X = X[:, ::-1] 
```

```python
scaler=StandardScaler()
scaler.fit(X)
X=scaler.transform(X)
```

```python
plt.figure(figsize=(5,3))
plt.scatter(X[:, 0], X[:, 1], s=40, cmap='plasma')
plt.show()
```

```python
centroida1=X[sample(range(1,400),2)[0]]
```

```python
centroida2=X[sample(range(1,400),2)[1]]
```

```python
def cluster_label_func(X,centroida1,centroida2):
    cluster_label=[]
    for i in range(len(X)):
        d1=math.dist(X[i],centroida1) 
        d2=math.dist(X[i],centroida2) 
        if(d1<d2):
            cluster_label.append(1)
        else:
            cluster_label.append(2)
            
    return(cluster_label)
```

```python

plt.figure(figsize=(18,15))

plt.subplot(3,3,1)
plt.scatter(X[:, 0], X[:, 1], s=40, cmap='plasma')
centroide = np.array([centroida1,centroida2])
colors = ['red','red']
plt.scatter(centroide[:,0],centroide[:,1],marker='o',c = colors)


plt.subplot(3,3,2)        
plt.scatter(X[:, 0], X[:, 1], c=cluster_label_func(X,centroide[0],centroide[1]), s=40, cmap='plasma')


plt.subplot(3,3,3)        
plt.scatter(X[:, 0], X[:, 1], c=cluster_label_func(X,centroide[0],centroide[1]), s=40, cmap='plasma')
centroide = np.array([np.mean(X[np.where(np.array(cluster_label_func(X,centroide[0],centroide[1]))==1)], axis=0),np.mean(X[np.where(np.array(cluster_label_func(X,centroide[0],centroide[1]))==2)], axis=0)])
colors = ['red','red']
plt.scatter(centroide[:,0],centroide[:,1],marker='o',c = colors)

plt.subplot(3,3,4)        
plt.scatter(X[:, 0], X[:, 1], c=cluster_label_func(X,centroide[0],centroide[1]), s=40, cmap='plasma')

plt.subplot(3,3,5)        
plt.scatter(X[:, 0], X[:, 1], c=cluster_label_func(X,centroide[0],centroide[1]), s=40, cmap='plasma')
centroide = np.array([np.mean(X[np.where(np.array(cluster_label_func(X,centroide[0],centroide[1]))==1)], axis=0),np.mean(X[np.where(np.array(cluster_label_func(X,centroide[0],centroide[1]))==2)], axis=0)])
colors = ['red','red']
plt.scatter(centroide[:,0],centroide[:,1],marker='o',c = colors)

plt.subplot(3,3,6)        
plt.scatter(X[:, 0], X[:, 1], c=cluster_label_func(X,centroide[0],centroide[1]), s=40, cmap='plasma')

plt.subplot(3,3,7)        
plt.scatter(X[:, 0], X[:, 1], c=cluster_label_func(X,centroide[0],centroide[1]), s=40, cmap='plasma')
centroide = np.array([np.mean(X[np.where(np.array(cluster_label_func(X,centroide[0],centroide[1]))==1)], axis=0),np.mean(X[np.where(np.array(cluster_label_func(X,centroide[0],centroide[1]))==2)], axis=0)])
colors = ['red','red']
plt.scatter(centroide[:,0],centroide[:,1],marker='o',c = colors)

plt.subplot(3,3,8)        
plt.scatter(X[:, 0], X[:, 1], c=cluster_label_func(X,centroide[0],centroide[1]), s=40, cmap='plasma')

plt.subplot(3,3,9)        
plt.scatter(X[:, 0], X[:, 1], c=cluster_label_func(X,centroide[0],centroide[1]), s=40, cmap='plasma')
centroide = np.array([np.mean(X[np.where(np.array(cluster_label_func(X,centroide[0],centroide[1]))==1)], axis=0),np.mean(X[np.where(np.array(cluster_label_func(X,centroide[0],centroide[1]))==2)], axis=0)])
colors = ['red','red']
plt.scatter(centroide[:,0],centroide[:,1],marker='o',c = colors)



plt.show()
```
